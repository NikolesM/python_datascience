```python
#pip install matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creamos dos arrays para visualizar en un grafico
month = np.array(['E','F','M','A','MA'])
sales = np.array([20,25,30,28,35])

#configuramos el tamaño del grafico
plt.figure(figsize = (8,6))

#crear el grafico 
plt.plot(month,sales,marker ='o', color = 'blue')
plt.title('Ventas mensuales de un producto')
plt.xlabel('Meses')
plt.ylabel('Ventas en miles de unnidades')
plt.show()

```

```python
#Creando un grafico de dispersión
#configuramos el tamaño del grafico
plt.figure(figsize = (8,6))

hours = [2,3,4,5,6,7,8]
exam = [55,60,65,70,75,80,85]

plt.scatter(hours,exam,color='green')
plt.title('Relación entre horas estudiadas y puntaje')
plt.xlabel('Horas')
plt.ylabel('Puntaje')
plt.show()
```

#Personalización de graficos

```python
#grafico de dispersion entre dos variables
plt.figure(figsize = (8,6))

hours = [2,3,4,5,6,7,8]
exam_score_student1 = [55,60,65,70,75,80,85]
exam_score_student2 = [50,58,63,69,74,78,83]
#creamos grafico de dispersión de dos estudiantes
plt.scatter(hours,exam_score_student1,marker = 'o', linestyle= '-',label = 'Estudiante 1',  color='green')
plt.scatter(hours,exam_score_student2,marker = 's', linestyle= '--',label = 'Estudiante 2',  color='red')
plt.title('Relación entre horas estudiadas y puntaje de dos estudiantes')
plt.xlabel('Horas')
plt.ylabel('Puntaje')
plt.show()
```

```python
#grafico de lineas entre dos variables
plt.figure(figsize = (8,6))

hours = [2,3,4,5,6,7,8]
exam_score_student1 = [55,60,65,70,75,80,85]
exam_score_student2 = [50,58,63,69,74,78,83]
#creamos grafico de dispersión de dos estudiantes
plt.plot(hours,exam_score_student1,marker = 'o', linestyle= '-',linewidth = 2,label = 'Estudiante 1',  color='green')
plt.plot(hours,exam_score_student2,marker = 's', linestyle= '--',linewidth = 2, label = 'Estudiante 2',  color='red')
plt.title('Relación entre horas estudiadas y puntaje de dos estudiantes')
plt.xlabel('Horas')
plt.ylabel('Puntaje')
plt.legend(loc='upper left', fontsize=10, frameon=True)
plt.show()
```

# Bars and pies
## Bars

```python
plt.figure(figsize=(8,6))
categories = ['Producto A','Producto B','Producto C']
sales = [120,150,90]

#creacion del grafico de barras verticales
plt.bar(categories,sales, color = 'skyblue', label='Ventas mensuales')

#anotación con flecha para destacar un punto especifico
plt.annotate('Maximo de ventas',xy=('Producto B',150), xytext=('Producto C',160),  arrowprops=dict(facecolor='black', shrink = 0.05))

plt.title('Ventas de productos en un mes')
plt.xlabel('Productos')
plt.ylabel('ventas')
#plt.legend()

plt.show()
```

```python
plt.figure(figsize=(8,6))
categories = ['Producto A','Producto B','Producto C']
sales = [120,150,90]

#creacion del grafico de barras horizontales
plt.barh(categories,sales, color = 'skyblue', label='Ventas mensuales')

plt.title('Ventas de productos en un mes')
plt.xlabel('Ventas')
plt.ylabel('Productos')
#plt.legend()

plt.show()
```

## Pies

```python
plt.figure(figsize=(8,6))
products = ['Producto A','Producto B','Producto C']
market_sahere = [50,35,15]

#creacion del grafico de pastel
plt.pie(market_sahere, labels=products,autopct = '%1.1f%%', startangle=140,shadow=True,
        colors=['gold','lightcoral','orange'])

plt.title('Participación de mercado por producto')


plt.axis('equal')
#plt.legend()

plt.show()
```

# Box

```python
#Podemos establecer una 'semilla' es decir un rango de valores de inicio lo que permitiria que a pesar que los valores son random se mantengan
np.random.seed(0)
ages = [np.random.normal(30,5,100),
        np.random.normal(40,5,100),
        np.random.normal(35,5,100)]
print(ages)
```

```python
#graficamos los valores anteriores con un grafico 'box' de cajas

plt.boxplot(ages, patch_artist=True, notch = True,vert= True,
            labels = ['Group1','Group2','Group3'])
plt.title('Distribución de edades por grupo')
plt.xlabel('Grupo')
plt.ylabel('Edad (años)')
plt.show()
```

# Series de tiempo
* Qué son: Son datos recolectados o registrados en secuencias cronológicas, donde cada punto de datos tiene una marca temporal. Ejemplos incluyen temperaturas diarias, precios de acciones o registros de ventas.
* Para qué sirven: Se utilizan para analizar y prever tendencias, patrones estacionales y ciclos a lo largo del tiempo. Son esenciales para el análisis predictivo, el monitoreo de eventos y la toma de decisiones basadas en datos históricos.

```python
#importaciones de librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter #importamos el modulo para dar formato a las fechas

dates =pd.date_range(start = '2023-01-01', periods=100) #generamos un rango de fechas que empiece desde 2023 y con periodo o frecuencia 100

values = np.random.rand(100).cumsum() #generamos 100 valores aleatorios y con la funcion cumsum generamos una suma acumulada a esos valores

data= pd.DataFrame({'Date': dates, 'Values': values})



```

```python
#crear grafico de lineas
fig, ax = plt.subplots(figsize =(12,6))

ax.plot(data['Date'], data['Values'], color = 'green')

plt.xticks(rotation=45)

plt.title('Serie de tiempo con formato en las fechas')
plt.xlabel('Date')
plt.xlabel('Valor')
plt.show()
```

```python
dates = pd.date_range(start = '2023-01-01', periods = 12, freq = 'M')
sales = np.random.randint(1000,5000,size = 12)

sales_data = pd.DataFrame({'Date': dates, 'Sales': sales})

plt.plot(sales_data['Date'],sales_data['Sales'], marker='o', linestyle = '-', label = 'Ventas mensuales') #personalizamos el grafico agregando marcador y estilo

#damos formato de fecha en texto
plt.gca().xaxis.set_major_formatter(DateFormatter('%b %Y'))

plt.xticks(rotation = 45) #damos rotación a los labels
plt.title('Analisis de ventas mensuales')
plt.xlabel('Date')
plt.xlabel('Sales') 
plt.legend() #agregamos leyenda
plt.tight_layout()
plt.show()
```

# Gridspec

Una herrramienta que permite crear layouts personalizados y subplots de diferentes tamñaos y formas
* figure es el término general que se usa para referirse al objeto Figure en Matplotlib, que es el contenedor de todos los subgráficos y otros elementos gráficos.

* axes es el objeto Axes que representa el área del gráfico dentro de la figura.

* Subplots permiten organizar múltiples gráficos en una sola figura para facilitar la comparación.
* Layouts avanzados ofrecen una mayor flexibilidad en la disposición de los gráficos, permitiendo diseños personalizados y complejos para una visualización más efectiva.

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec

x = np.linspace(0,10,100) #linspace nos dice valor inicial, valor final, paso
y = np.sin(x) #creamos una variable que guarde el seno del valor de la variable x

data = np.random.randn(100)  #generamos 100 valores aleatorios
```

```python
gs = gridspec.GridSpec(2,2,height_ratios=[2,1], width_ratios=[1,1])
fig = plt.figure(figsize = (10,8)) #asignamos el tamaño de la figura
```

```python
gs = gridspec.GridSpec(2,2,height_ratios=[2,1], width_ratios=[1,1])
fig = plt.figure(figsize = (10,8)) #asignamos el tamaño de la figura

## primer subplot grande, ocupa toda la primera fila
ax1 = fig.add_subplot(gs[0,:])
ax1.plot(x,y, color = 'blue')
ax1.set_title('Seno de X')
ax1.set_xlabel('X')
ax1.set_ylabel('Sin(x)')

#Segundo subplot, ocupa la esquina inferior izquierda

ax2 = fig.add_subplot(gs[1,0])
ax2.hist(data,bins = 20, color = 'purple', edgecolor = 'black')
ax2.set_title('Histograma')
ax2.set_xlabel('Valor')
ax2.set_ylabel('Frecuencia')

#Tercer subplot, ubicado en la esquina inferior derecha
ax3 = fig.add_subplot(gs[1,1])
ax3.scatter(x,y, color = 'red', edgecolor = 'black')
ax3.set_title('Dispersión de Seno')
ax3.set_xlabel('X')
ax3.set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```
